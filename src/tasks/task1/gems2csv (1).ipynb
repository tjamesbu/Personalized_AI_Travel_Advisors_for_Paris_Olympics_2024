```python
def website2df(url):
  website_url = url
  tool = ScrapeWebsiteTool(website_url=website_url)
  result = tool.run()
  messages = prompt_template.format_messages(text=result)
  chat = ChatOpenAI(temperature=0.0, model=llm_model)
  response = chat(messages)
  new_text = response.content[7:-3]
  new_list = eval(new_text)
  df = pd.DataFrame(new_list)
  return df

```
