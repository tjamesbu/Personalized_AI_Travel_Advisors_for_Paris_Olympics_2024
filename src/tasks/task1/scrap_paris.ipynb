```python

# Import necessary libraries
import praw
import pandas as pd
import json

# Initialize Reddit API
reddit = praw.Reddit(client_id='your reddit client id',
                     client_secret='your_reddit_secret_key',
                     user_agent='username')

# Define the subreddit and search query
subreddit = reddit.subreddit('ParisTravelGuide')
query = 'Hidden gems of Paris'

# Get relevant posts
posts = []
for submission in subreddit.search(query, sort='relevance', time_filter='all'):
    post_data = {
        "title": submission.title,
        "post_score" : submission.score,
        "total_comments" : submission.num_comments,
        "url" : submission.url,
        "comments" : [comment.body for comment in list(submission.comments)]
    }
    posts.append(post_data)

# Save data to a JSON file
with open("hidden_gems_of_paris.json", "w") as json_file:
    json.dump(posts, json_file, indent=4)

print("Data saved to hidden_gems_of_paris.json")
```
