## Creating Agents

````python
cultural_difference_expert = Agent(
    role="cultural difference expert",
    goal="Identify key clutural differnces between a given country and paris to  "
         "understand the dispointement of the tourist ",
    backstory=(
        "You are a cultural anthropologist and as a,"
        "member of big mental health advisor agnecy for supporting tourists "
        "who suffers from Paris sundrom,"
        "you analyze differences in Local Laws and Customs, "
        "Language and Communication, Social Norms and Behaviors "
        "Your work is crucial in paving the way "
        "for raising cultural awarness and calming the tourists."
    ),
    allow_delegation=False,
    verbose=True
)
````

````python
general_mental_health_advisor = Agent(
    role="Mental Health advisor",
    goal="To provide personalized, empathetic support and guidance to tourists experiencing Paris Syndrome",
    backstory=(
        "Within the mental health advisor agnecy for tourists, "
        "You stand out as the bridge between Paris Syndrome symptoms"
        "and the reorientation they need."
        "By creating conforting, personalized messages, "
        "you not only inform tourists about key cultural differnces and informations "
        "but also make them feel seen and heard and adjust their mental issues"
        "My role is pivotal in helping them adjust their mental state,"
        "alleviate their distress,"
        "and ultimately make their trip to Paris a much more"
        "enjoyable and fulfilling experience"
    ),
    allow_delegation=False,
    verbose=True
)
````

## Creating Tools

### crewAI Tools

````python
from crewai_tools import DirectoryReadTool, \
                         FileReadTool, \
                         SerperDevTool
````

````python
directory_read_tool = DirectoryReadTool(directory='./instructions')
file_read_tool = FileReadTool()
search_tool = SerperDevTool()
````

### Custom Tool
- Create a custom tool using crewAi's [BaseTool](https://docs.crewai.com/core-concepts/Tools/#subclassing-basetool) class

````python
from crewai_tools import BaseTool
````

- Every Tool needs to have a `name` and a `description`.
- For simplicity and classroom purposes, `SentimentAnalysisTool` will return `positive` for every text.
- When running locally, you can customize the code with your logic in the `_run` function.

````python
class SentimentAnalysisTool(BaseTool):
    name: str ="Sentiment Analysis Tool"
    description: str = ("Analyzes the sentiment of text "
         "to ensure positive and engaging communication.")
    
    def _run(self, text: str) -> str:
        # Your custom code tool goes here
        return "positive"
````

````python
sentiment_analysis_tool = SentimentAnalysisTool()
````

## Creating Tasks

- The culture_difference_task Task is using crewAI Tools.

````python
culture_difference_task = Task(
    description=(
        "Conduct an in-depth analysis of {country}, "
        "Utilize all available data sources "
        "to compile a detailed cultural comapraison, "
        "between the Paris and the {country}"
        "that potentailly cause cultural shock for a tourist coming with our offerings. "
        "from this {country} and visiting Paris"
        "Don't make assumptions and "
        "only use information you absolutely sure about."
    ),
    expected_output=(
        "A comprehensive report on {country}, "
        "including local Laws and Customs, "
        "language and Communication, Social Norms and Behaviors"
        "key cultural differnces, "
    ),
    tools=[directory_read_tool, file_read_tool, search_tool],
    agent=cultural_difference_expert,
)
````

- The general_health_advisor_task is using a  custom Tool `SentimentAnalysisTool`, as well as crewAI's `SerperDevTool` (search_tool).

````python
general_health_advisor_task = Task(
    description=(
        "Using the insights gathered from "
        "the cultural difference between paris and the country report on {country}, "
        "and using the the tourist issues concerning Paris Syndrom {tourist_issues_description}"
        "craft a personalized message "
        "The message should address their issues "
        "and help them reframe their expectations,"
        "process the cultural shock and disappointment,"
        "making them understanding the key differnces between paris and {country}"
        "consider only diferences that may cause a shock"
        "and how to cope with each aspect"
        "Your communication must resonate "
        "demonstrating a deep understanding of "
        "their concerns and needs.\n"
        "Don't make assumptions and only "
        "use information you absolutely sure about."
    ),
    expected_output=(
        "A personalized guidance message tha "
    ),
    tools=[sentiment_analysis_tool, search_tool],
    agent=general_mental_health_advisor,
)
````

## Creating the Crew

````python
crew = Crew(
    agents=[cultural_difference_expert, 
            general_mental_health_advisor],
    
    tasks=[culture_difference_task, 
           general_health_advisor_task],
	
    verbose=2,
	memory=True
)
````

## Running the Crew

**Note**: LLMs can provide different outputs for they same input, so what you get might be different than what you see in the video.

````python
inputs = {
    "country": "Tunisia",
    "tourist_issues_description": "I had such high expectations for Paris, but the reality was completely different from what I imagined. I feel disillusioned and disappointed. The city is much dirtier, noisier, and more crowded than I expected. The people seem rude and unhelpful. I'm having trouble communicating and getting around",
   
}

result = crew.kickoff(inputs=inputs)
````
